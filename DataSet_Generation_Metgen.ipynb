```python
def create_emotion_dataset(data, emotion_list, input_img_path, out_path_text, out_path_img,folder='train'):
  """
  Function to create dataset for training/validation/testing. Takes in
  emotion list and dataframe containing image hashes for yfcc dataset from 
  ParlAI display command.
  """
  # Created filtered dataset
  ndata = data[data['personality'].isin(emotion_list)]
  ndata.reset_index(drop=True,inplace=True)
  # Create paths 
  if not os.path.exists(out_path_text):
    os.makedirs(out_path_text)
  if not os.path.exists(out_path_img):
    os.makedirs(out_path_img)
  # Get list of image hashes
  img_list = ndata['image_hash'].tolist()
  # Get images and save it in the demarcated folder 
  for img_hsh in tqdm(img_list):
    try:
      img = Image.open(input_img_path + img_hsh +'.jpg')
      img.load()
      img_array = np.asarray(img, dtype='int32')
      imsave( out_path_img + img_hsh + ".jpg" , img_array)
    except:
      print("Issue with image hash: " + str(hash))
      continue
  
  ndata.to_csv(out_path_text+ folder+ "_caption.csv")
```
